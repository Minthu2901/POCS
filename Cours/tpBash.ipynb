# 1 Afficher le dossier courant
```bash
$ pwd
```
# 2 Remonter dans le répertoire /home/vscode
```bash
$ cd ../../../home/vscode #chemain relatif
$ cd /home/vscode         #chemain absolue
```
# 3 Créer un dossier miniTP dans ce répertoire 
```bash
$ mkdir miniTP
```
# 4 Afficher le contenu de tous les fichiers ou dossiers présents
```bash
$ ls -al
```
# 5 Créer un fichier vide notes.txt
```bash
$ touch ./notes.txt
```
# 6 Ajouter dans ce fichier la ligne "Linux is powerful"
```bash
$ nano ./notes.txt #puis on écrit la ligne souhaiter
$ echo "Linux is powerful" > ./notes.txt #met le "" en remplacant l'ancient text
$ echo "Linux is powerful" >> ./notes.txt #met le "" sans remplacer
#le > marche aussi sur d'autre commande (tree >> essai.txt ==> ajoute le tree dans le fichier text)
```
# 7 Renommer ce fichier sans le déplacer en notes_backup.txt
```bash
$ mv ./notes.txt ./notes_backup.txt
```
# 8 Copier ce fichier dans le dossier cours de votre workspace
```bash
$ cp ./notes_backup.txt /workspaces/POCS/Cours/notes_backup.txt
```
# 9 Compter le nombre de mots dans le fichier
```bash
$ wc ./notes_backup.txt
```
# 10 Afficher la/les lignes avec le mot powerfull dans la console
```bash
$ grep powerful ./notes.txt
```
